```python
import random

def word(min_word_length):
    
    """This Function takes min_word_length as Parameter and return Random Word from 'data.txt' file,
    which legth is greater or equal to Player Selected Length(Min Word Length)"""
    
    w_word = []
    temp = open('data.txt','r')
    words = temp.read().split()
    for word in words:
        if len(word) >= min_word_length:
            w_word.append(word.lower())
    return random.choice(w_word)

def min_word_length(name):
    
    """This Function takes Player Name as Parameter and take the input of min_word_length Player want to Guess,
    in range 4-11 and return min_word_length"""
    
    while True:
        min_word_length = input(f"{name} Minimum Word Length Do You Want Guess in Range[4-11] : ")
        try:
            min_word_length = int(min_word_length)
            if 4 <= min_word_length <= 11:
                return min_word_length
            else:
                print(f'{min_word_length} is not betwen 4 to 11. Try again..')
        except ValueError:
            print(f'{min_word_length} is not an integer. Try again..')
    
def min_attempts(name):
    
    """This Function takes Player name as Parameter and take the input of min_attempts Player Required to Guess,
    and return min_attempts"""
    
    while True:
        attempts = input(f'{name} Minimum Attempts Do you want in Range[1-22] : ')
        try:
            attempts = int(attempts)
            if 1 <= attempts <= 22:
                return attempts
            else:
                print(f'{attempts} is not betwen 4 to 11. Try again..')
        except ValueError:
            print(f'{attempts} is not an integer. Try again..')

def inp():
    
    """This Function takes input of name from Player and call the Function word,min_word_length and min_attempts,
    and return the name,guess_word,attempts(with accepting condition:attempts>unique character in word)."""
    
    name=input("Enter Yor Name : ").strip().upper()
    print(f'Good Luck {name}!')
    word_length = min_word_length(name)
    guess_word = word(word_length)
    attempts = min_attempts(name)
    if len(set(guess_word)) > attempts:
        d = len(set(guess_word)) - attempts
        attempts += random.randint(d+2,d+7)
    else:
        attempts += random.randint(1,5)
    print(f'{name} You got {attempts} attempts.')        
    return name,guess_word,attempts
    
def game():
    
    """This Function call the function inp and takes the guess from player and return the result"""
    
    name,guess_word,attempts=inp()
    temp1 = ''
    incorrect = ''
    while attempts > 0:
        temp2 = 0
        temp = input('Enter chr : ')
        if len(temp) == 1:
            temp1 += temp
            for i in guess_word:
                if i in temp1:
                    print(i,end='')
                else:
                    print('-',end='')
                    temp2+=1
            print()
            if temp2 == 0:
                print(f"Congrats {name}, You Won")
                print(f'Correct Word : {guess_word}')
                return 1
                break
            attempts -= 1
            if temp not in guess_word:
                incorrect += temp
                print(f'Incorrects : {incorrect}')
                print("wrong!")
            print(f'{name} you have chances {attempts} left.')

            if attempts == 0:
                print("you Lose")
                print(f'Correct Word : {guess_word}')
                return 0              
                    
        elif len(temp) > 1:
            print('Please, Enter only one character ')
game()
while True:
    pl = input('Do You Want to play Again [y/Y] : ').strip()
    if pl[0].lower() == 'y':
        game()
    else:
        print('Thank You For Playing.')
        break
```
